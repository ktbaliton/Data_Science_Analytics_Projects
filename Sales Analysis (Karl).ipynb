```python
#Use .apply()

all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1] + ' ' + x.split(',')[2].split(' ')[1])

all_data.head()
```

#### Bonus

```python
#Use .apply()

def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

#Using function
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
```

### Question 4: What products are most often sold together?

```python
#Return rows by boolean value; 'Keep' to mark True or False
df = all_data[all_data['Order ID'].duplicated(keep=False)]

#Get each 'Product' and join
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
#https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)
```

#### Based on quantity

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
quantity_ordered.head()
```

```python
products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical')
plt.ylabel('Quantity Sold')
plt.xlabel('Product')
#plt.grid()
plt.show
```

#### Product Prices

```python
prices = all_data.groupby('Product').mean()['Price Each']

print(prices)
```

#### Adding Prices in the Quantity Sold per Product

```python
import numpy as np

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered)
ax2.plot(products, prices, 'r-')

ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Sold')
ax2.set_ylabel('Price (USD)', color='r')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
